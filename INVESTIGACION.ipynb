#### Empezamos importanto librerias standar

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import userdata
```

```python
import requests

API_TOKEN = userdata.get('CLASH_API')
BASE_URL = 'https://api.clashroyale.com/v1'
HEADERS = {
    'Authorization': f'Bearer {API_TOKEN}',
    'Accept': 'application/json'
}

def get_player_info(player_tag):
    formatted_tag = player_tag.replace('#', '%23')
    url = f"{BASE_URL}/players/{formatted_tag}"

    response = requests.get(url, headers=HEADERS)

    if response.status_code == 200:
        return response.json()
    else:
        return f"Error: {response.status_code} - {response.text}"

tag = "#8P8CLL2JL"
player_data = get_player_info(tag)
```

##### obtenermos las primeras 5 cartas del jugador

```python
print("Cartas principales del jugador:")
i = 0
for card in player_data['cards']:
    print(f"- {card['name']} (Nivel: {card['level']})")
    i += 1
    if i == 5:
        break
```
